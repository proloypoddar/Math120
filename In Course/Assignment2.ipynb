1. $\int \sin (a x) e^{-b x} d x$




```python
from sympy import *
import matplotlib.pyplot as plt
import numpy as np

x, a, b = symbols('x a b')
integrand = sin(a*x) * exp(-b*x)
func = lambdify(x, integrand.subs([(a, 1), (b, 1)]), 'numpy')
x_vals = np.linspace(-10, 10, 1000)
y_vals = func(x_vals)
plt.plot(x_vals, y_vals)
plt.show()

integrand = sin(a*x) * exp(-b*x)
integral = integrate(integrand, x)
integral = simplify(integral)
pprint( integral)

a_val = 1
b_val = 1
n = 1000
dx = 20 / n
area = 0
for i in range(n):
    x_val = -10 + i * dx
    area += dx * integrand.subs([(a, a_val), (b, b_val), (x, x_val)]).evalf()
print(f"The rectangular method approximation with n={n} is: {area}")

a_val = 1
b_val = 1
n = 100000
x_min = -10
x_max = 10
y_min = 0
y_max = max(y_vals) 

area = 0
for i in range(n):
    x_val = np.random.uniform(x_min, x_max)
    y_val = np.random.uniform(y_min, y_max)
    if y_val <= func(x_val):
        area += 1

area_ratio = area / n
integral = area_ratio * (x_max - x_min) * (y_max - y_min)

print(f"The Monte Carlo approximation with n={n} is: {integral}")
```


2. $\int_0^{\infty} \frac{\sin ^{-1}(2 x)}{1+x^2} d x$


```python
from sympy import *
import numpy as np
import matplotlib.pyplot as plt

x = symbols('x')
integrand = asin(2*x)/(1+x**2)
integral = integrate(integrand, (x, 0, oo))

print(f"The exact value of the integral is: {(integral)}")
func = lambdify(x, integrand, 'numpy')
x_vals = np.linspace(0, 10, 1000)
y_vals = func(x_vals)
plt.plot(x_vals, y_vals)
plt.show()

n = 1000
dx = 10 / n
area = 0
for i in range(n):
    x_val = i * dx
    area += dx * integrand.subs(x, x_val).evalf()
print(f"The rectangular method approximation with n={n} is: {area}")

n = 100000
x_min = 0
x_max = 10
y_min = 0
y_max = max(y_vals)

area = 0
for i in range(n):
    x_val = np.random.uniform(x_min, x_max)
    y_val = np.random.uniform(y_min, y_max)
    if y_val <= func(x_val):
        area += 1

area_ratio = area / n
integral = area_ratio * (x_max - x_min) * (y_max - y_min)

print(f"The Monte Carlo approximation with n={n} is: {integral}")
```

3. $\int_0^{2 \pi} \frac{1}{(a-\cos (x))^2+(b-\sin (x))^2} d x$


```python
from sympy import *
import numpy as np
import matplotlib.pyplot as plt

x = symbols('x')
a, b = symbols('a b', real=True)
integrand = 1 / ((a-cos(x))**2 + (b-sin(x))**2)
integral = integrate(integrand, (x, 0, 2*pi))
print(f"The exact value of the integral is: {(integral)}")

func = lambdify(x, integrand, 'numpy')
x_vals = np.linspace(0, 2*pi, 1000)
y_vals = func(x_vals)
plt.plot(x_vals, y_vals)
plt.show()

n = 1000
dx = 2*pi / n
area = 0
for i in range(n):
    x_val = i * dx
    area += dx * integrand.subs([(a, 2), (b, 3)]).subs(x, x_val).evalf()
print(f"The rectangular method approximation with n={n} is: {area}")

n = 100000
x_min = 0
x_max = 2*pi
y_min = 0
y_max = max(y_vals) 

area = 0
for i in range(n):
    x_val = np.random.uniform(x_min, x_max)
    y_val = np.random.uniform(y_min, y_max)
    if y_val <= func(x_val):
        area += 1

area_ratio = area / n
integral = area_ratio * (x_max - x_min) * (y_max - y_min)

print(f"The Monte Carlo approximation with n={n} is: {integral}")
```

5. $\int_{-1}^2 x e^{6 x} d x$
